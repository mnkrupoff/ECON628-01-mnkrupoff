```python
import pandas as pd
```

```python
bl=pd.read_stata("/Users/mkrupoff/Dropbox/CSI_IFPRI_Thesis/data/India_baseline.dta",convert_categoricals=False)
```

```python
def drop_columns(l,u, drops, df):
    column_drops=range(l,u)
    df.drop(df.columns[[column_drops]], axis=1, inplace=True)
    return df
```

```python
drop_columns(1,16, column_drops, bl)
drop_columns(3,10, column_drops, bl)
drop_columns(8,10, column_drops, bl)
drop_columns(13,17, column_drops, bl)
```
